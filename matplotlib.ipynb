```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(1,19,2) #arange prints numbers of that range in array 
x
```

```python
y=[-3,22,5.6,36,-22,15,6.99,89,26.66] #y has all types in a list
type(y)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("speed") # xlabel & ylabel - labels x & y axes
plt.ylabel("dist")
plt.title("sample") # heading
plt.show()
```

```python
plt.ylabel("sample") # empty graph since not plotted
```

```python
y1=[-36,22,5.6,36,-2.2,15,6.99,-6,26.66] # defining y1 & y2
y2=[-3,2.2,5.6,36,-22,158,6.99,89,26.66]
print(x)
plt.plot(x,y1,linestyle="dashdot",linewidth=3,label="1") 
plt.plot(x,y2,label="2",color="r",) 
#ADDING COORDINATES TO POINTS
for i, j in zip(x, y1):
  plt.text(i, j, '({}, {})'.format(i, j)) #iterator objects that will be joined together
for a, b in zip(x, y2):
  plt.text(a, b, '({}, {})'.format(a, b))
plt.xlabel("speed")
plt.ylabel("dist")
plt.title("sample")
plt.legend()
plt.show()
```

```python
plt.plot(y1)
print(y1)
print(x)
plt.show()
```

```python
plt.plot(x,y1,marker="o",ms=10,mec="r",mfc="k") # ms for markersize,mec for edge color,mfc for face color
plt.show
```

```python
plt.bar(x,y1) #bar plot
plt.bar(x,y2)
y1,y2
```

```python
#here plt.show() seperates single graph into 3 components
plt.bar(x,y2,label="1")
plt.legend()
plt.show()
plt.bar(x,y1)
plt.show()
plt.bar(x,y1)
plt.bar(x,y2)
plt.show()
```

```python
n=10+np.random.randn(100) # FOR HIST
n
```

```python
x=np.arange(1,200,2) # Or x=[x for x in range(len(n))]
x
```

```python
plt.barh(x,n,height=3)
```

```python
plt.hist(n,label="1",edgecolor="k",bins=3) #hist plots based on frequency, bins change width of bars
plt.legend()
n.max()-n.min()/10
```

```python
plt.hist(n,label="1",edgecolor="k",bins=20,cumulative=-1)
plt.show()
plt.hist(n,label="1",edgecolor="k",bins=20,cumulative=1)
plt.show()
```
